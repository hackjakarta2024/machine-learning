```python
from collections import Counter

def get_embedding(text, model="text-embedding-3-large"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input=[text], model=model).data[0].embedding

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def determine_preferences(user_data, df):
    """
    Determines liked categories and preferred cuisines based on user's historical data.
    """
    past_orders = user_data.get('past_orders', [])

    if not past_orders:
        return [], []

    # Create embeddings for past orders
    past_order_embeddings = [get_embedding(order) for order in past_orders]

    # Compute similarities with all items in the DataFrame
    df['embedding'] = df['combined'].apply(lambda x: get_embedding(x))
    df['similarity'] = df['embedding'].apply(lambda x: max(cosine_similarity(x, order_emb) for order_emb in past_order_embeddings))

    # Find the most similar items
    similar_items = df.sort_values('similarity', ascending=False).head(len(past_orders))

    # Aggregate liked categories and preferred cuisines
    liked_categories = Counter()
    preferred_cuisines = Counter()

    for _, row in similar_items.iterrows():
        categories = row['resto_cat'].split(',')
        for category in categories:
            liked_categories[category.strip()] += 1

        cuisines = row['title'].split()
        for cuisine in cuisines:
            preferred_cuisines[cuisine.strip()] += 1

    # Get the most common categories and cuisines
    top_liked_categories = [cat for cat, _ in liked_categories.most_common(3)]
    top_preferred_cuisines = [cuisine for cuisine, _ in preferred_cuisines.most_common(3)]

    return top_liked_categories, top_preferred_cuisines
```

```python
# Step 1: Generate the Dataset
num_users = 100
np.random.seed(42)
user_ids = range(1, num_users + 1)
ages = np.random.randint(18, 60, size=num_users)
incomes = np.random.randint(100000, 1500000, size=num_users)
avg_meals_ordered = np.random.randint(1, 10, size=num_users)
# Generate last order days ago for the last three orders
total_order_last1d = np.random.randint(1, 6, size=num_users)
food_names = [
    'Nasi Goreng', 'Sate Ayam', 'Rendang', 'Gado-Gado',
    'Bakso', 'Mie Goreng', 'Ayam Penyet', 'Soto',
    'Nasi Uduk', 'Pempek', 'Martabak', 'Gudeg'
]
# Promos: each promo is given with a probability of 20%
promo_probs = [0.2] * 10
promos = {f'promo_{i+1}': np.random.choice([0, 1], size=num_users, p=[1 - p, p]) for i, p in enumerate(promo_probs)}

# Add control group: 20% of users get no promo
control_group_size = int(num_users * 0.2)
control_indices = np.random.choice(num_users, control_group_size, replace=False)

for promo in promos:
    promo_data = promos[promo]
    promo_data[control_indices] = 0  # Ensure control group does not receive any promo
    promos[promo] = promo_data

# Transaction probability influenced by promos and food preferences
base_prob = 0.1
promo_influence = sum(promos.values()) * 0.05  # Each promo increases the transaction probability by 0.05
transact_prob = base_prob + promo_influence  # + food_influence
transact = np.random.binomial(1, transact_prob.clip(0, 1))

# Create DataFrame
df = pd.DataFrame({
    'user_id': user_ids,
    'age': ages,
    'income': incomes,
    'avg_meals_ordered': avg_meals_ordered,
    'top3_latest_orders': [', '.join(random.sample(food_names, 3)) for _ in range(num_users)],
    'total_order_last1d': total_order_last1d,
    'transact': transact
})

for promo_name, promo_data in promos.items():
    df[promo_name] = promo_data

df
```

```python
import pandas as pd
import numpy as np
from econml.dr import LinearDRLearner
from sklearn.linear_model import LinearRegression
from sklearn.metrics.pairwise import cosine_similarity
from scipy.optimize import linprog
from sklearn.ensemble import RandomForestClassifier
import joblib
```

```python
# Step 2: Promo Impact Analysis
Y = df['transact'].values
X = df[['age', 'income', 'avg_meals_ordered', 'total_order_last1d']].values
promo_effects = {}

for promo in promos:
    T = df[promo].values
    learner = LinearDRLearner(model_regression=LinearRegression(),
                              model_propensity=RandomForestClassifier())
    learner.fit(Y=Y, T=T, X=X)
    promo_effects[promo] = learner.effect(X)

promo_effects_df = pd.DataFrame(promo_effects)
promo_effects_df
```

```python
# Save the trained models
for promo, learner in promo_effects.items():
    joblib.dump(learner, f'{promo}_learner.pkl')

# # Load the trained models
# loaded_promo_effects = {}
# for promo in promo_effects.keys():
#     loaded_promo_effects[promo] = joblib.load(f'{promo}_learner.pkl')
```

```python
# Calculate the uplift compared to control group
control_impact = np.mean(df.loc[control_indices, 'transact'])
promo_uplifts = {promo: effects - control_impact for promo, effects in promo_effects.items()}

promo_uplift_df = pd.DataFrame(promo_uplifts)
```

```python
# Step 4: Define Promo Costs and Budget
promo_costs = {
    'promo_1': 20000,   # Cost for 30% discount promo
    'promo_2': 30000,   # Cost for Buy 1 Get 1 Free
    'promo_3': 15000,   # Cost for 20% discount on first order
    'promo_4': 10000,   # Cost for Free delivery
    'promo_5': 25000,   # Cost for 10% cashback
    'promo_6': 10000,   # Cost for Rp10,000 off
    'promo_7': 20000,   # Cost for Free drink
    'promo_8': 15000,   # Cost for 15% discount for new users
    'promo_9': 5000,    # Cost for Rp5,000 off for referring a friend
    'promo_10': 20000   # Cost for Free dessert
}
total_budget = 10000000  # Example budget of Rp10,000,000
```

```python
# Step 5: Calculate Estimated Gains from Promos
average_transaction_value = 50000  # Example average value of Rp50,000
promo_gains = promo_uplift_df * average_transaction_value
user_ids = df['user_id'].values
promo_names = list(promo_costs.keys())
promo_gains_array = promo_gains.values.flatten()
promo_costs_array = np.array([promo_costs[promo] for promo in promo_names] * num_users)
promo_allocation_matrix = np.zeros((num_users, num_users * len(promo_names)))

for i in range(num_users):
    promo_allocation_matrix[i, i*len(promo_names):(i+1)*len(promo_names)] = 1
```

```python
# Step 6: Optimization Using Linear Programming
c = -promo_gains_array
A_ub = [promo_costs_array]
b_ub = [total_budget]
A_eq = promo_allocation_matrix
b_eq = np.ones(num_users)
bounds = [(0, 1)] * len(promo_gains_array)

result = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

if result.success:
    optimal_allocation = result.x.reshape(num_users, len(promo_names))
else:
    print("Optimization failed:", result.message)
    optimal_allocation = np.zeros((num_users, len(promo_names)))  # Default to no promo allocation if optimization fails
```

```python
# Step 7: Combine Results and Generate Final Recommendations
recommended_promos = [promo_names[np.argmax(user_alloc)] if np.max(user_alloc) > 0 else 'no_promo' for user_alloc in optimal_allocation]

final_recommendations = df[['user_id', 'top3_latest_orders']].copy()
final_recommendations['recommended_promo'] = recommended_promos
```
